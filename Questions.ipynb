```python
# Question 2

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for the Tesla revenue data
url = "https://finance.yahoo.com/quote/TSLA/financials"  # Update this if necessary

# Send a GET request to fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

revenue_data = []
for row in soup.find_all("div", class_="D(tbr)"):
    cells = row.find_all("span")
    data_row = [cell.get_text() for cell in cells]
    if data_row:  # Only add non-empty rows
        revenue_data.append(data_row)

tesla_revenue = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])

print(tesla_revenue.tail())
```

```python
# Question 5

import matplotlib.pyplot as plt
import yfinance as yf

# Download Tesla stock data
tesla_data = yf.download("TSLA", start="2010-01-01", end="2023-01-01")
tesla_data.reset_index(inplace=True)

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.xlabel("Date")
    plt.ylabel("Close Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot the Tesla stock data with a title
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
